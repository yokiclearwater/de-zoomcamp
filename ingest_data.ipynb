## Testing Connection

```python
from sqlalchemy import create_engine

engine  = create_engine('postgresql://postgres:postgres@db:5432/ny_taxi')
```

```python
print(pd.io.sql.get_schema(green_trip_data, name='green_tripdata', con=engine))
```

```python
import time
from datetime import timedelta

def timeit(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print(f'Execution time: {timedelta(seconds=end-start)}')
        return result
    return wrapper
```

```python
green_trip_data['store_and_fwd_flag']
```

```python
@timeit
def process_chunks():
    trip_data_iter = pd.read_csv('data/green_tripdata_2019-10.csv', iterator=True, chunksize=100000, low_memory=False)
    
    total_rows = 0
    chunk_number = 0
    
    while True:
        try:
            chunk_start = time.time()
            
            chunk = next(trip_data_iter)
            chunk['lpep_pickup_datetime'] = pd.to_datetime(chunk['lpep_pickup_datetime'])
            chunk['lpep_dropoff_datetime'] = pd.to_datetime(chunk['lpep_dropoff_datetime'])
            
            # Insert chunk into database
            chunk.to_sql(name='green_taxi_data', con=engine, if_exists='append', index=False)
            
            chunk_end = time.time()
            chunk_time = timedelta(seconds=chunk_end-chunk_start)
            
            # display(chunk.head())
            chunk_number += 1
            total_rows += len(chunk)
            
            print(f'Chunk {chunk_number}:')
            print(f'- Shape: {chunk.shape}')
            print(f'- Memory usage: {chunk.memory_usage().sum() / 1024 / 1024:.2f} MB')
            print(f'- Running total rows: {total_rows}')
            print(f'- Chunk processing time: {chunk_time}\n')
            
            # Query Row From Database
            query = 'SELECT COUNT(*) FROM green_taxi_data'
            row_count = pd.read_sql_query(query, engine)
            print(f'Total rows in database: {row_count.iloc[0,0]}')
            
        except StopIteration:
            print("Finished processing all chunks")
            break
            
    return total_rows

total_rows = process_chunks()
print(f'\nTotal rows processed: {total_rows}')
```
